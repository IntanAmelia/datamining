# **Decision Tree**

```python
import numpy as np 

import pandas as pd 

import sklearn

from sklearn.model_selection import train_test_split

from sklearn.metrics import confusion_matrix 

from sklearn.tree import DecisionTreeClassifier 

from sklearn.metrics import accuracy_score 

from sklearn.datasets import load_iris

from matplotlib import pyplot as plt

from sklearn import datasets

from sklearn import tree
```

```python
#Loading datasets 

data = "https://raw.githubusercontent.com/IntanAmelia/datamining/main/dataR2.csv"

df = pd.read_csv(data)
df
```

```python
X = df[['Age',	'BMI',	'Glucose',	'Insulin',	'HOMA',	'Leptin',	'Adiponectin',	'Resistin',	'MCP.1']]
y = df['Classification']

print("X : ",X)

print("Y : ",y)
```

```python
# Splitting the dataset into train and test 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

```python
# Decision tree classifier 

clf= DecisionTreeClassifier(random_state = 100)

#fitting the training data

clf.fit(X_train, y_train)
```

```python
# prediction on random data

X=[[70,	20.8,	110,	6.77,	1,	4.8,	17,	5.8,	289.334]]

Y_pred=clf.predict(X)

print(Y_pred)

# prediction on X_test (testing data )

Y_pred=clf.predict(X_test)

print(Y_pred)
```

```python
#Tree plotting 

tree.plot_tree(clf)
```
